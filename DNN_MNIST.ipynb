```python
#############################################################
#A simple DNN with 2 hidden layers on MNIST
#############################################################
import tensorflow as tf
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
n_input=28*28
n_output=10
n_hidden1=300
n_hidden2=100
mnist=input_data.read_data_sets("/tmp/data/")
```

```python
X_training, y_training=mnist.train.images, mnist.train.labels
X_testing, y_testing=mnist.test.images, mnist.test.labels
```

```python
def hidden_layer(X, name, n_neurons, activation=None):
    with tf.name_scope(name):
        n_input=int(X.get_shape()[1])
        stddev=2/np.sqrt(n_input)
        init=tf.truncated_normal((n_input, n_neurons), stddev=stddev)
        W=tf.Variable(init, name='Weights')
        b=tf.Variable(tf.zeros([n_neurons]), name="Bias")
        z=tf.matmul(X,W)+b
        if activation=='relu':
            return tf.nn.relu(z)
        else:
            return z
```

```python
##################################################
#Construction: two hidden layers, activation is relu
##################################################
X=tf.placeholder(tf.float32, shape=(None,n_input), name='X')
y=tf.placeholder(tf.int64, shape=(None), name='y')

with tf.name_scope("dnn"):
    hidden_1=hidden_layer(X, "hidden1", n_hidden1, activation="relu")
    hidden_2=hidden_layer(hidden_1, "hidden2", n_hidden2, activation="relu")
    logits=hidden_layer(hidden_2, "outputs", n_output)
```

```python
######################################################
#Define loss
#######################################################
with tf.name_scope("loss"):
    xentropy=tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)
    loss=tf.reduce_mean(xentropy, name="loss")
    
#########################################################
#Optimizer
#########################################################
learning_rate=0.01
with tf.name_scope("train"):
    optimizer=tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
    training_op=optimizer.minimize(loss)
##########################################################
#Accuracy
##########################################################
with tf.name_scope("accuracy"):
    correct=tf.nn.in_top_k(logits, y, 1)
    accuracy=tf.reduce_mean(tf.cast(correct, tf.float32))
```

```python
##############################################################
#Executiuon
##############################################################
n_epoche=400
batch_size=100
init=tf.global_variables_initializer()

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epoche):
        for iteration in range(X_training.shape[0]//batch_size):
            X_batch, y_batch=X_training[iteration*batch_size:(iteration+1)*batch_size]\
                                ,y_training[iteration*batch_size:(iteration+1)*batch_size]
            sess.run(training_op, feed_dict={X:X_batch, y:y_batch})
        training_accuracy=accuracy.eval(feed_dict={X:X_training, y:y_training})
        testing_accuracy=accuracy.eval(feed_dict={X:X_testing, y:y_testing})
        
        print("Epoch: ", epoch, "training accuracy: ", training_accuracy, "testing accuracy: ", testing_accuracy)
        
```
